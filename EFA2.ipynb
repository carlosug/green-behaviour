##### Q1.  Data cleaning and transformation (1.5 points)
- a) Import the dataset (csv file) using pandas into a dataframe named, `df`


- b) Transform the DataFrame `df` from Likert scale to numbers. You should use the following mapping for the transformation:

```python
import pandas as pd
url = 'input/data.csv'
df = pd.read_csv(url)
df.shape
```

```python
questions = ["isocntry", "q2_1", "q4_1", "q4_2", "q4_3", "q4_4", "q4_5", "q4_6", "q4_7", "q4_8", "q10", "q5", "q6"]
df_final = df.filter(questions)
df_final
```

```python
df_transformed = df_final.rename(columns={'q2_1': 'att1', 'q4_1': 'att2', 'q4_2': 'att3','q4_3': 'mo1', 'q4_4': 'pcb1', 'q4_5': 'mo2', 'q4_6': 'att4', 'q4_7': 'sn1', 'q4_8': 'pcb2', 'q10': 'pcb3', 'q5': 'pcb4', 'q6': 'wtp'})
col_means = df_transformed.mean()
col_means
```

```python
df_transformed = df_transformed.fillna(col_means)
df_transformed.att1.values
```

```python
from factor_analyzer import calculate_kmo
kmo_all,kmo_model=calculate_kmo(df_transformed)
kmo_model
```

```python
no_of_items = df_transformed.shape[1] # we have 13 questions = number of columns = the second parameter of df.shape
no_of_items
```

```python
from factor_analyzer import FactorAnalyzer
fa = FactorAnalyzer(n_factors=no_of_items, method='principal') # set number of factors to the number of items and method ='principal'
fa.fit(df_transformed)
```

```python
ev, v = fa.get_eigenvalues()
ev
```

```python
import matplotlib.pyplot as plt

# x-axis should print the number of factors (= no. of items so from 1 to 8)
#Use the range function to get numbers from 1 to 8. Check documentation of range() here: https://www.w3schools.com/python/ref_func_range.asp
x_axis = range(1,no_of_items+1,1) 

# y-axis should contain the eigen values
y_axis = ev


#Documentation: https://matplotlib.org/3.5.1/plot_types/index.html
        
#First make a scatter plot (just plot the points without the line)
plt.scatter(x_axis,y_axis)

#Then make a line plot connecting these points
plt.plot(x_axis,y_axis)

#Add a plot title
plt.title('Scree Plot')

# Add the lable for x-axis
plt.xlabel('Factors')

# Add the lable for y-axis
plt.ylabel('Eigenvalue')

#Add the grid lines (optional)
plt.grid()

#Show the plot
plt.show()
```

```python
no_of_retained_factors = 3
fa = FactorAnalyzer(n_factors = no_of_retained_factors , method='principal', rotation ='varimax') #default rotation is set to promax (oblique rotation)
fa.fit(df_transformed) 
```

```python
fa.loadings_
```

```python
df_loadings = pd.DataFrame(fa.loadings_,index=df_transformed.columns, columns =['Factor 0','Factor 1','Factor 2'])
df_loadings.style.background_gradient(cmap = "winter")
```
