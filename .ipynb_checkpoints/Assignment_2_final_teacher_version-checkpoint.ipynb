##### Q1.  Data cleaning and transformation (1.5 points)
- a) Import the dataset (csv file) using pandas into a dataframe named, `df`


- b) Transform the DataFrame `df` from Likert scale to numbers. You should use the following mapping for the transformation:
    
    
    Strongly Agree ---> 6

    Agree ---> 5

    Somewhat Agree ---> 4
    
    Somewhat Disagree ---> 3

    Disagree ---> 2

    Strongly Disagree ---> 1
    
    The name of the transformed dataframe should be `df_transformed`.
    
    

- c) Handle the missing values: Compute the median of each survey item/question and replace the missing values by the median of that particular item.

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/MaastrichtU-IDS/global-studies/main/semester4/tutorial6/data_assignment_2/tolerance_survey_assignment.csv'
df = pd.read_csv(url)
df
```

```python
df_transformed = df.replace(['Strongly Agree',
                   'Agree', 
                   'Somewhat agree',
                    'Somewhat disagree',
                   'Disagree', 
                   'Strongly Disagree'], [6,5,4,3,2,1])
df_transformed
```

```python
col_median = df_transformed.median()
col_median
```

```python
df_transformed = df_transformed.fillna(col_median)
df_transformed.q1.values
```

##### Q4. Conduct an Exploratory Factor analysis for the Tolerance Questionnaire ( 2.5 points)

- a) Conduct a Kaiser-Meyer-Olkin (KMO) test on the `df_transformed` dataframe (the dataframe should contain numeric values after transformation from the Likert scale) to evalaute if its suitable for Factor Analysis or not. Interpret the results of the KMO test.


- b) Conduct an Exploratory Factor Analysis on `df_transformed`. Set the number of factors `n_factors` equal to the number of items in the survey and choose `method = principal` for principal factor extraction method of exploratory factor analysis.


- c) Print the eigen values. Make the scree plot. 


- d) How many factors will you choose depending on the eigen values and the interpretation of the scree plot ? Explain the reason of your choice. (Use Markdown cell to answer) 


- e) What are the coordinates (x,y) of the point of inflexion ? (Use either Code or Markdown cell to answer).

```python
from factor_analyzer import calculate_kmo
kmo_all,kmo_model=calculate_kmo(df_transformed)
kmo_model
```

```python
no_of_items = df_transformed.shape[1] # we have 13 questions = number of columns = the second parameter of df.shape
no_of_items
```

```python
from factor_analyzer import FactorAnalyzer
fa = FactorAnalyzer(n_factors=no_of_items, method='principal') # set number of factors to the number of items and method ='principal'
fa.fit(df_transformed)
```

```python
ev, v = fa.get_eigenvalues()
ev
```

```python
import matplotlib.pyplot as plt

# x-axis should print the number of factors (= no. of items so from 1 to 8)
#Use the range function to get numbers from 1 to 8. Check documentation of range() here: https://www.w3schools.com/python/ref_func_range.asp
x_axis = range(1,no_of_items+1,1) 

# y-axis should contain the eigen values
y_axis = ev


#Documentation: https://matplotlib.org/3.5.1/plot_types/index.html
        
#First make a scatter plot (just plot the points without the line)
plt.scatter(x_axis,y_axis)

#Then make a line plot connecting these points
plt.plot(x_axis,y_axis)

#Add a plot title
plt.title('Scree Plot')

# Add the lable for x-axis
plt.xlabel('Factors')

# Add the lable for y-axis
plt.ylabel('Eigenvalue')

#Add the grid lines (optional)
plt.grid()

#Show the plot
plt.show()
```

```python
no_of_retained_factors = 3
fa = FactorAnalyzer(n_factors = no_of_retained_factors , method='principal', rotation ='varimax') #default rotation is set to promax (oblique rotation)
fa.fit(df_transformed) 
```

```python
fa.loadings_
```

```python
df_loadings = pd.DataFrame(fa.loadings_,index=df_transformed.columns, columns =['Factor 0','Factor 1','Factor 2'])
df_loadings.style.background_gradient(cmap = "winter")
```
